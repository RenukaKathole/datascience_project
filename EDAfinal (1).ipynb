```python
#1.Histogram plot
```

```python
sns.histplot(df['age'],kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
# Define feature columns (excluding target)
#prognosis: (0 or 1) (Target). It is 1 if the patient has retinopathy and 0 if he doesn't.
features = ['age', 'systolic_bp', 'diastolic_bp', 'cholesterol']
target = 'prognosis'
```

```python
for column in features:
    sns.histplot(df[column], kde=True, color='skyblue')
    plt.title(f'{column.capitalize()} Histogram')
    plt.show()
```

```python
#2.Boxplot
```

```python
for column in features:
    sns.boxplot(y=df[column], color='lightcoral')
    plt.title(f'{column.capitalize()} Boxplot')
    plt.show()
```

```python
# 3. Violin Plot
```

```python
for column in features:
    sns.violinplot(x=target, y=column, data=df, palette='Set3')
    plt.title(f'{column.capitalize()} vs {target.capitalize()} (Violin Plot)')
    plt.show()
```

```python
#4. Bar Plot (Mean by Target)
```

```python
for column in features:
    sns.barplot(x=target, y=column, data=df, palette='Set2')
    plt.title(f'Mean {column.capitalize()} by {target.capitalize()}')
    plt.show()
```

```python
#5. Regplot (vs Cholesterol)
```

```python
for column in features:
    if column != 'cholesterol':
        sns.regplot(x=df[column], y=df['cholesterol'], scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
        plt.title(f'{column.capitalize()} vs Cholesterol (Regplot)')
        plt.show()
```

```python
#6. Scatterplot
```

```python
sns.scatterplot(x="systolic_bp",y="diastolic_bp",hue="cholesterol",data=df)
```

```python
#7. Line Plot by Age (sorted)
```

```python
df_sorted = df.sort_values('age')
for column in features:
    sns.lineplot(x=df_sorted['age'], y=df_sorted[column], color='green')
    plt.title(f'{column.capitalize()} vs Age (Lineplot)')
    plt.show()
```

```python
#8. Heatmap
```

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])
corr = numeric_df.corr()
```

```python
sns.heatmap(corr, annot=True,cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
 #9. Pairplot
```

```python
sns.pairplot(df, hue=target)
plt.suptitle("Pair Plot of Features")
plt.show()
```

```python
#10. Pie Chart of Target Class
```

```python
class_counts = df[target].value_counts()
labels = ['No Retinopathy', 'Retinopathy']
plt.pie(class_counts, labels=labels, autopct='%1.1f%%', colors=['lightgreen', 'salmon'], startangle=140)
plt.title("Prognosis Class Distribution (Pie Chart)")
plt.axis('equal')
plt.show()
```
